## Datahantering
Datan som projektet baseras på är hämtat från kaggle.com där Martin Frederiksen sammanställt all relevant data kopplat till försäljningar av bostäder i Danmark.
Eftersom vi vill undersöka variablerna försäljningsdatum, hustyp, byggnationsår, antal rum, kvadratmeter, adress och stad behöver vi exkludera alla andra variabler för att kunna hantera datan på ett smidigt sätt. Vi kommer även endast titta på köpespriset så alla andra variabler om pris väljer vi att bortse från. Detta åstadkommes genom att sortera datan och exkludera irrelevanta variabler.
Så här ser den ursprungliga datan ut: 

```python
path = "C:/Users/Gustav/miniconda3/su-mt4007/Project/DKHousingPricesSample100k.csv"
data = pd.read_csv(path)
data
```

Efter vi har tagit bort irrelevanta variabler för vår frågeställning ser datan ut på följande sätt.

```python
cleared_data = data[["purchase_price","date", "house_type", "year_build", "no_rooms", "sqm", "address", "city"]]
cleared_data
```

## Medelvärdlig prisändring under åren
Vi skapar en graf som visar hur priserna generellt har ändrats under åren för bostäder i Danmark. Observera att stora prisskillnader bland bostäder gör att meddelvärdet inte är en överdrivet realistisk representation över hur priserna ändras eftersom att det är många kategorier som inte räknas in i denna analys som t.ex. hustyp, storlek osv. Vi tar endast meddelpriset av alla bostäder oavsett deras egenskaper.

```python
#  Konvertera "date" till datetime-format och extrahera året
copied_data = cleared_data.copy()
copied_data["date"] = pd.to_datetime(copied_data["date"])
copied_data["year"] = copied_data["date"].dt.year

# Filtrera data till endast år från och med 1992
copied_data = copied_data[copied_data["year"] >= 1992]

#  Beräkna och plotta genomsnittligt köppris per år
avg_price_per_year = copied_data.groupby("year")["purchase_price"].mean()

plt.figure(figsize=(12, 6))
avg_price_per_year.plot(kind="line", marker="o", color="blue", label="Genomsnittligt köpeskilling")
plt.title("Genomsnittligt köppris per år (1992 och framåt)")
plt.xlabel("År")
plt.ylabel("Genomsnittligt köppris (DKK)")
plt.grid(True)
plt.legend()
plt.show()
```

## Analys
Grafen visar en mycket tydlig indikation på att priserna för bostäder i Danmark generellt har stigit markant. Denna åskådliggörning berättar för oss att bostäders köpeskilling ökat från runt 0.5 miljoner danska kronor upp till 2.5 miljoner på 30 år. Denna ökning kan givetvis bero på många olika faktorer som t.ex. ändring i valutavärde, ändring i medborgares inkomst, vilka hustyper som är attraktiva på marknaden eller brist på bostäder. Vi kommer inte djupdyka i vad det är som påverkat prissändringen i detta projekt utan en demonstration av ändringen är vad vi vill åstadkomma.  

```python
#Beräkna och plotta genomsnittligt köppris per hustyp

avg_price_per_house_type = cleared_data.groupby("house_type")["purchase_price"].mean().sort_values()

plt.figure(figsize=(10, 6))
avg_price_per_house_type.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Genomsnittligt köppris per hustyp")
plt.xlabel("Hustyp")
plt.ylabel("Genomsnittligt köppris (DKK)")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

# 7. Valfritt: Utför ytterligare analyser, t.ex. hur antal rum påverkar priset
avg_price_per_rooms = cleared_data.groupby("no_rooms")["purchase_price"].mean().sort_index()

plt.figure(figsize=(10, 6))
avg_price_per_rooms.plot(kind="bar", color="orange", edgecolor="black")
plt.title("Genomsnittligt köppris per antal rum")
plt.xlabel("Antal rum")
plt.ylabel("Genomsnittligt köppris (DKK)")
plt.xticks(rotation=0)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

## Analys
I dessa två stapeldiagram illustreras kostnaden av olika hustyper samt hur hus med flera rum generellt påverkar priset. Detta ger ingen indikation på trender i när vissa bostäder är dyrare och billigare under tidsgången utan endast en generell sammanfattning under de 30 åren i fråga. Vi ser att sommarstugor tenderar att vara billigast med ett medelpris på ungefär 1.2 miljoner DKK medans gårdar säljs för över 2.5 miljoner DKK. 

I diagram två ser vi skillnaden på priser för bostäder med olika antal rum och rimligtsvis är summan generellt stigande ju fler rum en bostad har även om det inte alltid tycks vara en markant skillnad. Speciellt kan vi konstatera att en bostad med ett rum säljs för ungefär 1.2 miljoner medans en bostad med 15 rum säljs för 6.2 miljoner kronor vilket är en ökning på 5 miljoner. Detta tyder på ett tydligt samband mellan antal rum och pris.

## Korrelation mellan variabler och pris
För att ta reda på vilka variabler som påverkar priset mest gör vi nu en korrelationsplot för de numeriska variablerna. 

```python
new_data = cleared_data.copy()
# Extrahera år från "date"
new_data["year"] = pd.to_datetime(new_data["date"], errors="coerce").dt.year

# Omvandla kategoriska variabler ("house_type", "city") till dummy-variabler
new_data = pd.get_dummies(new_data, columns=["house_type", "city"], drop_first=True)

# Ta endast numeriska kolumner för att undvika fel
numeric_data = new_data.select_dtypes(include=["number"])

# Beräkna korrelation med "purchase_price"
correlations = numeric_data.corr()["purchase_price"].sort_values(ascending=False).drop("purchase_price")

#  Plotta korrelationerna
plt.figure(figsize=(12, 8))
correlations.plot(kind="barh", color="skyblue", edgecolor="black")
plt.title("Korrelation mellan 'purchase_price' och andra variabler")
plt.xlabel("Korrelation")
plt.ylabel("Variabler")
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.show()
```

## Analys
Den variabel som har högst korrelation av dem vi undersöker är vilket år bostaden sålts, och detta är ett rimligt resultat eftersom att vi såg i grafen tidigare att priserna ökat mycket med åren. Det som påverkar bostädernas priser mest är därmed faktorer som ändrats med tiden såsom valutans värde, utbud på bostäder etc.

Någonting intressant från korrelationsplottarna är att byggnationsår har negativ korrelation med köpespris vilket innebär att pris går ner för senare byggnationsår, detta skulle kunna bero på att vissa gamla bostäder har väldigt hög efterfrågan och är attraktiva på marknaden och får därmed höga värden. Om man hade undersökt korrelationen mellan priset och hur gammal bostaden är hade vi förmodligen fått en positiv korrelation. 

Vi ser även att det finns tydlig korrelation för kvadratmeter och antal rum med priset och det är rimligt eftersom större bostäder tenderar att vara dyrare.
